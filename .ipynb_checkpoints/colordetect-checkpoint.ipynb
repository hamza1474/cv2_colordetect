# Code

```python
#green color range
start = (15,25,25)
end = (75,255,255)
```

```python
lo_square = np.full((10, 10, 3), start, dtype=np.uint8) / 255.0
do_square = np.full((10, 10, 3), end, dtype=np.uint8) / 255.0
```

```python
from matplotlib.colors import hsv_to_rgb
plt.subplot(1, 2, 1)
plt.imshow(hsv_to_rgb(do_square))
plt.subplot(1, 2, 2)
plt.imshow(hsv_to_rgb(lo_square))
plt.show()
```

```python
def mask_image(image):
    img_hsv = cv2.imread(os.path.join('images/',image), cv2.COLOR_BGR2HSV)
    plt.imshow(img_hsv)
    mask = cv2.inRange(img_hsv, start, end)
    result = cv2.bitwise_and(img_hsv, img_hsv, mask=mask)
    return mask, result
```

```python
def display_image(mask, result):    
    
    plt.subplot(1, 2, 1)
    plt.imshow(mask, cmap='gray')
    plt.subplot(1, 2, 2)
    plt.imshow(result)
    plt.show()
```

```python
mask, result = mask_image(images[9])
```

```python
display_image(mask, result)
```

```python
#SAVE OUTPUT TO DISK
from PIL import Image
output=np.concatenate((img_hsv, result))
im = Image.fromarray(output)
im.save("file.jpeg")
```
